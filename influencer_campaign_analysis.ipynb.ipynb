```python
import matplotlib.pyplot as plt

plt.boxplot(df["Spend_USD"])
plt.title("Campaign Spend")
plt.show()
```

```python
df["Engagement_Rate"] = (
    (df["Likes"] + df["Comments"] + df["Shares"])
    / df["Reach"]
) * 100
```

```python
df["ROI"] = (
    (df["Revenue_USD"] - df["Spend_USD"])
    / df["Spend_USD"]
) * 100
```

```python
df["Conversion_Rate"] = (
    df["Conversions"] / df["Reach"]
) * 100
```

```python
df.to_csv("cleaned_influencer_campaign_dataset.csv", index=False)
```

```python
conversion_rate = (
    df["Conversions"].sum() /
    df["Reach"].sum()
) * 100

print("Conversion Rate:", round(conversion_rate,2), "%")
```

```python
kpis = {
    "Total Campaigns": df["Campaign_ID"].nunique(),
    "Total Reach": df["Reach"].sum(),
    "Total Revenue": df["Revenue_USD"].sum(),
    "Total Spend": df["Spend_USD"].sum(),
    "Average Engagement Rate (%)": round(df["Engagement_Rate"].mean(),2),
    "Average ROI (%)": round(df["ROI"].mean(),2),
    "Cost Per Engagement": round(df["Spend_USD"].sum()/df["Engagements"].sum(),2),
    "Conversion Rate (%)": round((df["Conversions"].sum()/df["Reach"].sum())*100,2)
}

for key, value in kpis.items():
    print(f"{key}: {value}")
```

```python
def influencer_type(followers):
    if followers < 10000:
        return "Nano Influencer"
    elif followers < 100000:
        return "Micro Influencer"
    elif followers < 1000000:
        return "Macro Influencer"
    else:
        return "Mega Influencer"

df["Influencer_Type"] = df["Followers"].apply(influencer_type)
```

```python
df["Influencer_Type"].value_counts()
```

```python
def roi_category(roi):
    if roi >= 100:
        return "High ROI"
    elif roi >= 50:
        return "Medium ROI"
    else:
        return "Low ROI"

df["ROI_Category"] = df["ROI"].apply(roi_category)
```

```python
df["ROI_Category"].value_counts()
```

```python
def engagement_category(rate):
    if rate >= 10:
        return "High Engagement"
    elif rate >= 5:
        return "Medium Engagement"
    else:
        return "Low Engagement"

df["Engagement_Category"] = df["Engagement_Rate"].apply(engagement_category)
```

```python
df["Engagement_Category"].value_counts()
```

```python
df["Cost_Per_Engagement"] = df["Spend_USD"] / df["Engagements"]

def cost_efficiency(cpe):
    if cpe < 1:
        return "Cost Efficient"
    elif cpe < 2:
        return "Moderately Efficient"
    else:
        return "Expensive"

df["Cost_Category"] = df["Cost_Per_Engagement"].apply(cost_efficiency)
```

```python
roi_type = df.groupby("Influencer_Type")["ROI"].mean()
print(roi_type)
```

```python
cost.plot(kind="bar", figsize=(8,5))

plt.title("ROI by Cost Efficiency")
plt.ylabel("Average ROI")

plt.show()
```
